```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

import warnings
warnings.filterwarnings("ignore")
```

```python
da=pd.read_csv('application_record.csv')
da
```

```python
dc=pd.read_csv('credit_record.csv')
dc
```

* ## merging the two data sets on column ID to not Coflict the data 
***

```python
data = pd.merge(da, dc, on="ID")
data
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
data['STATUS'].value_counts()
```

* ### we turning the target column to 0 and 1 on the below condition 
0: 1-29 days past due 1: 30-59 days past due 2: 60-89 days overdue 3: 90-119 days overdue 4: 120-149 days overdue 5: Overdue or bad debts, write-offs for more than 150 days C: paid off that month X: No loan for the month

```python
data['STATUS'].replace(['C','X'], [1,1],inplace=True)
target = data
target.loc[target['STATUS'] != 1, 'STATUS'] = 0
target['STATUS']= pd.to_numeric(target['STATUS'])
print(target['STATUS'])
# we see ther is no palance in the data 
```

* ## replace the opject values to numirec
****

```python
data['CODE_GENDER'].replace(['F','M'], [0,1],inplace=True)
data['FLAG_OWN_CAR'].replace(['Y','N'], [1,0],inplace=True)
data['FLAG_OWN_REALTY'].replace(['Y','N'], [1,0],inplace=True)
```

```python
data.nunique()
```

```python
# drop column 'FLAG_MOBIL' cuse of the fixed value
data.drop(columns=['FLAG_MOBIL'],inplace =True)
```

```python
data['OCCUPATION_TYPE'].value_counts()
```

```python
# drop all NAN values to not disperse the prediction and get wrong values
data.dropna(axis=0,inplace=True)
```

```python
# hanc we have a lage amuont of data i use get_dummies function to have equality bettwen mempers
data = pd.get_dummies(data, columns=
               ['NAME_EDUCATION_TYPE', 'NAME_INCOME_TYPE', 'NAME_FAMILY_STATUS'
               , 'NAME_HOUSING_TYPE', 'OCCUPATION_TYPE'])
for col in data:
    data[col]=pd.to_numeric(data[col])
```

```python
data.info()
```

```python
# see the correlatoin with heat map
plt.figure(figsize=(40,40))
sb.heatmap(data.corr(),annot=True)
plt.show()
# ther is no corrleation bettwen the target and the data 
```

* ### split the data into train and test 
***

```python
x=data.drop(['STATUS'],axis=1).values
y=data.loc[:,'STATUS'].values
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.35)
```

* ### applaing the machine learning model Classification 
***

```python
from sklearn.preprocessing import StandardScaler

sd= StandardScaler()
train_x=sd.fit_transform(train_x)
test_x=sd.transform(test_x)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(train_x,train_y)
```

```python
y_pred=model.predict(test_x)
from sklearn.metrics import confusion_matrix
print (confusion_matrix(y_pred,test_y))
```

```python
from sklearn.metrics import accuracy_score,recall_score,f1_score,precision_score
print ('Accuracy : ', accuracy_score(test_y,y_pred))
print ('Recall : ',recall_score(test_y,y_pred))
print ('precision : ',precision_score(test_y,y_pred))
print ('F1Score : ',f1_score(test_y,y_pred))
```
